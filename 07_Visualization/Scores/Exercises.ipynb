### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame that should look like the one below.

```python
df = pd.DataFrame(
    columns=['first_name', 'last_name', 'age', 'preTestScore', 'postTestScore','female'],
    data=[
        ['Jason', 'Miller', 42, 4, 25,False],
        ['Molly', 'Brown', 23, 20, 14,True],
        ['Tina', 'Lee', 28, 22, 35,True],
        ['Jake', 'Miller', 24, 2, 62,False],
        ['Amy', 'Cooze', 73, 3, 70,True]
    ]
)

df.head()
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.scatterplot(x='preTestScore', y='postTestScore', data=df, size='age')
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.scatterplot(x='preTestScore', y='postTestScore', data=df, size='age', hue='female')
```
